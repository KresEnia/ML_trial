```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv")
df
```

```python
df.columns
```

```python
x = df[['age', 'anaemia', 'creatinine_phosphokinase', 'diabetes', 'high_blood_pressure', 'platelets',
       'serum_creatinine', 'serum_sodium', 'sex', 'smoking', 'time']]
y = df['ejection_fraction']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regression': DecisionTreeClassifier(random_state=42),
    'Random Forest Regression': RandomForestClassifier(random_state=42)
}
```

```python
predictions = {}
for model_name, model in models.items():
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    predictions[model_name] = y_pred
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"Model: {model_name}")
    print(f"Mean squared error: {mse:.2f}")
    print(f"R squared: {r2:2f}")
    print("=" * 50)
```

```python
plt.figure(figsize=(12, 8))
for model_name, y_pred in predictions.items():
    sns.scatterplot(x=y_test, y=y_pred, label=model_name)

plt.xlabel('Actual Ejection fraction')
plt.ylabel('Predicted Ejection fraction')
plt.title('Model Predictions vs. Actual Values')
plt.legend()
plt.show()
```

```python
#Feature importance
for model_name, model in models.items():
    if "Tree" in model_name:
        feature_importances = model.feature_importances_
        feature_names = x.columns
        importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})
        importance_df = importance_df.sort_values(by='Importance', ascending=False)

        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=importance_df)
        plt.title(f'{model_name} - Feature importance for Ejection Fraction Prediction')
        plt.show()
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
coefficients = linear_model.coef_
feature_names = x.columns
coefficients_df = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
coefficients_df = coefficients_df.sort_values(by='Coefficient', ascending=False)

print("Linear Regression Coefficients:")
coefficients_df
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Coefficient', y='Feature', data=coefficients_df)
plt.title('Linear Regression Coefficients for Ejection Fraction Prediction')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ConfLongDemo_JSI.txt", header = None)
df
```

```python
column_names = [
    'sequence', 'Tag identificator', 'timestamp', 'date FORMAT', 'x coordinate', 'y coordinate',
    'z coordinate', 'activity']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
x_clustering = df[['x coordinate', 'y coordinate', 'z coordinate']]
```

```python
scaler_clustering = StandardScaler()
x_clustering_scaled = scaler_clustering.fit_transform(x_clustering)
```

```python
kmeans = KMeans(n_clusters=6, random_state=42)
```

```python
df['cluster'] = kmeans.fit_predict(x_clustering_scaled)
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='x coordinate', y='y coordinate', hue='cluster', data=df, palette='viridis')
plt.title('Clustering Results')
```

```python
#dimensionality reduction
df.info()
```

```python
from sklearn.decomposition import PCA
x_dim_reduction = df[['x coordinate', 'y coordinate', 'z coordinate']]
```

```python
scaler_dim_reduction = StandardScaler()
x_scaled_dim_reduction = scaler_dim_reduction.fit_transform(x_dim_reduction)
pca = PCA(n_components=2)
```

```python
x_pca = pca.fit_transform(x_scaled_dim_reduction)
```

```python
x_pca
```

```python
df['PCA_component_1'] = x_pca[:, 0]
df['PCA_component_2'] = x_pca[:, 1]
```

```python
df.head()
```

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='PCA_component_1', y='PCA_component_2', hue='activity', data=df, palette='viridis')
plt.title('PCA for Dimensionality Reduction Results for activity data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Activity', fontsize='small', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
